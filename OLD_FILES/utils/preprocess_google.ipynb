```python
!pip install google-cloud-storage pandas
!pip install pyarrow fsspec gcsfs
import pandas as pd
import pyarrow.parquet as pq
import pandas as pd

base_url = 'gs://nix-cl-dataset-gcs/ungrouped_output_p95_'
extension = '.parquet'

urls = [base_url + str(i).zfill(12) + extension for i in range(23)]
print(urls)
```

```python
file_urls = urls

# Initialize an empty list to store individual DataFrames
dfs = []

# Load each file into a DataFrame and append it to the list
for file_url in file_urls:
    print(file_url)
    df_temp = pd.read_parquet(file_url, engine='pyarrow')
    dfs.append(df_temp)

# Concatenate all DataFrames into a single DataFrame
df = pd.concat(dfs, ignore_index=True)
```
